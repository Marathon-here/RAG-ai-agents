```python
#PDF loader
from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader('/content/attention.pdf')
textdocs = loader.load()
textdocs
```

```python
pip install langchain_text_splitters
```

```python
#split the textual data into chucks for LLM model
from langchain_text_splitters import CharacterTextSplitter
text_splitter=CharacterTextSplitter(separator="\n\n",chunk_size=50,chunk_overlap=20) #chunk_size ==> divides the data into 100 charcters size of paragraphs,
                                                                                    #chunk_overlap ==> overlapping betwwen one para to other paragraph data
splitteddocument = text_splitter.split_documents(textdocs)
splitteddocument #this finaldocumnets have the chunks of data
```

```python
%pip install langchain_huggingface
```

```python
#embedding the huggingface model to text document
#embedding means converting splitted textual to vectors
from langchain_huggingface import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
embeddings
```

```python
query_result = embeddings.embed_query("finaldocumnet") #the chunks are convertto vectors
query_result
```

```python
#storing the vector into database
from langchain.vectorstores import FAISS
db = FAISS.from_documents(splitteddocument,embeddings) #agruments are splitted chucks data and vectors
db
```
