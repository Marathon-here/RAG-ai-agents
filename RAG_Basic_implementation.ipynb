```python
from langchain_community.document_loaders import TextLoader #using textloader class we will load data from text file
loader = TextLoader('/content/speech.txt') #copy the path from colab files
print(loader) #check the data address
text_document = loader.load()
for doc in text_document: #reading the line by line from file through loader
  print(doc)
```

```python
#install text_splitters from langchain
pip install langchain_text_splitters
```

```python
#split the textual data into chucks for LLM model
from langchain_text_splitters import CharacterTextSplitter
text_splitter=CharacterTextSplitter(separator="\n\n",chunk_size=10,chunk_overlap=0) #chunk_size ==> divides the data into 100 charcters size of paragraphs,
                                                                                    #chunk_overlap ==> overlapping betwwen one para to other paragraph data
finaldocument = text_splitter.split_documents(text_document)
finaldocument #this finaldocumnets have the chunks of data

```

```python
#embedding the huggingface model to text document
#embedding means converting splitted textual to vectors
from langchain_huggingface import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
embeddings
```

```python
#Web Based Loader
from langchain_community.document_loaders import WebBaseLoader #webbaseloder class is used to load the webpage data
import bs4    #
loader = WebBaseLoader(web_path=('https://lilianweng.github.io/posts/2022-06-09-vlm/',), #path of website
                       bs_kwargs=dict(parse_only= bs4.SoupStrainer(class_=("post-title","post-content","post-header"))  #soupstrainer is used yo extract the in webbaseloader
                       ))

docs = loader.load()
docs
```
